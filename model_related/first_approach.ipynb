```python
import pandas as pd
import glob

# Get a list of all files in the folder
all_files = glob.glob('scrapers/all_seasons_data/2017-2018/gw_0/*.csv')

# Read the first file into a DataFrame
df = pd.read_csv(all_files[0])

# For each remaining file
for file in all_files[1:]:
    # Read the file into a DataFrame
    df_other = pd.read_csv(file)
    if 'goalkeepers' in file:
        continue
    # Merge the DataFrame with the existing DataFrame on the 'player' column
    df = pd.merge(df, df_other, on=['player','shirtnumber','position','nationality','age'],how='inner',suffixes=('', '_remove'))

# Remove duplicate rows
df = df.drop_duplicates()
df = df.reset_index()
# Print the first 5 rows of the merged DataFrame
df.head()
```
