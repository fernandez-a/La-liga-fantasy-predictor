### Outliers

Previamente se ha visualizado los outliers, se utilizará el rango intercuartílico iqr para ello. Q3 - Q1 (75%-25%)

```python
Q1_tp = filtered_data['tp'].quantile(0.25)
Q3_tp = filtered_data['tp'].quantile(0.75)
IQR_tp = Q3_tp - Q1_tp
```

```python
limite_inferior = Q1_tp - (1.5 * IQR_tp)
limite_superior = Q3_tp + (1.5 * IQR_tp)
```

```python
outliers_tp = (filtered_data['tp'] < limite_inferior) | (filtered_data['tp'] > limite_superior)

filtered_data_no_tp_outliers = filtered_data[~outliers_tp]
```

```python
filtered_data_no_tp_outliers.to_csv('all_seasons_no_outliers.csv',index=False)
```

```python
plt.figure(figsize=(8, 6))

sns.boxplot(x='position', y='tp', data=filtered_data_no_tp_outliers)
plt.title('Distribución del total de puntos por posición')
plt.xlabel('Posición Jugador')
plt.ylabel('Puntos totales')

plt.savefig('./figures/boxplot_tp_no_outliers.png')

plt.show()
```

### Correlation Analysis

```python
filtered_data_no_tp_outliers = filtered_data_no_tp_outliers.drop(columns = 'position')
```

```python
summary_stats = filtered_data_no_tp_outliers.describe()

correlation_matrix = filtered_data_no_tp_outliers.corr()


plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Matriz correlación")

plt.savefig('./figures/correlation.png')

plt.show()
```

Se can 

```python
num_matchweeks = [3,5,10]

grouped = filtered_data_no_tp_outliers.sort_values(['Temporada', 'Jornada']).groupby('playerId')

rolling_avg_columns = ['mins', 'g', 'ga', 'oaa', 'pae', 'br', 'tsa', 'wc', 'pla']

rolling_avgs = []
for num_matchweeks in [3, 5, 10]:
    rolling_avg = grouped.rolling(window=num_matchweeks, min_periods=1).mean().reset_index(level=0, drop=True)
    rolling_avg = rolling_avg[rolling_avg_columns]
    rolling_avg.columns = [f'avg_{col}_last{num_matchweeks}w' for col in rolling_avg_columns]
    rolling_avgs.append(rolling_avg)

rolling_avg_all = pd.concat(rolling_avgs, axis=1)


efficiency_metrics = filtered_data.copy()
efficiency_metrics['goals_per_min'] = efficiency_metrics['g'] / efficiency_metrics['mins'].replace(0, 1)
efficiency_metrics['assists_per_min'] = efficiency_metrics['ga'] / efficiency_metrics['mins'].replace(0, 1)
efficiency_metrics['recoveries_per_game'] = efficiency_metrics['br'] / efficiency_metrics['mins'].replace(0, 1) * 90

enhanced_data = pd.concat([filtered_data_no_tp_outliers, rolling_avg_all, efficiency_metrics[['goals_per_min', 'assists_per_min', 'recoveries_per_game']]], axis=1)

enhanced_data.head()
```

Se puede observar un aumento de la correlación gracias a las nuevas variables que han sido creadas

```python
correlation_matrix = enhanced_data.corr()


threshold = 0.30
highly_correlated_features = correlation_matrix.index[abs(correlation_matrix["tp"]) > threshold]

subset_corr_matrix = enhanced_data[highly_correlated_features].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(subset_corr_matrix, fmt=".2f", cmap='coolwarm')
plt.title("Heatmap of Highly Correlated Features with 'tp'")
plt.show()
```
